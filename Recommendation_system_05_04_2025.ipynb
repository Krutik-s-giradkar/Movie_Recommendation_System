```python
import ast
ast.literal_eval('[{"name": "Adventure"},{"Fantasy"},{"id":28, "name" : "Action"}]')
```

```python
def convert1(text):
  ext1=[]
  counter = 0
  for i in ast.literal_eval(text):
    if counter < 5:
      ext1.append(i['name'])
    counter += 1

  return ext1
```

```python
movies['cast'] = movies['cast'].apply(convert1)
```

```python
movies['cast'][0]
```

```python
def fetch_director(text):
  ext2 = []
  for i in ast.literal_eval(text):
    if i['job'] == 'Director':
      ext2.append(i['name'])
  return ext2
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(ext):
  ext4 = []
  for i in ext:
    ext4.append(i.replace(" ",""))
  return ext4
```

# Feature Engineering / Extraction Method - Bag of Words (BOW)

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
vector = cv.fit_transform(new['context']).toarray()
vector.shape
```

```python
pd.DataFrame(vector).head()
```

```python
cv.get_feature_names_out()
```

# Building Recommendataton syaytem basis Similarity Matrix (cosin-similarity)

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
new[new['title'] == 'John Carter'].index[0]
```

```python
len(new['title'])
```

```python
def recommendationsystem(movies):
  index = new[new['title']==movies].index[0]
  distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector:vector[1])
  for i in distance[1:6]:
    print(new.iloc[i[0]].title)
```

```python
recommendationsystem('Avatar')
```

```python
recommendationsystem('John Carter')
```

```python
recommendationsystem('Spider')
```

```python
recommendationsystem('Spider-Man')
```
